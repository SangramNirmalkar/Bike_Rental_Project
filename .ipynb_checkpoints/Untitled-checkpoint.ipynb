Dataset contains both numerical and categorical features. Some columns are incorrectly stored as object type due to presence of invalid values like '?'.

Found '?' values in multiple columns, confirming data inconsistency. These need to be handled before further analysis.

# Data cleaning

```python
# replacing '?' values with nan

df.replace('?', np.nan, inplace=True)
```

Replaced invalid '?' entries with NaN to standardize missing value representation.

```python
df.isnull().sum()
```

Identified missing values across multiple columns including temperature, humidity, and user counts.

```python
# dropping rows where casual and registered == nan
df = df.dropna(subset=['casual','registered'])
```

Rows with missing casual or registered values were dropped because they directly contribute to the target variable (cnt). Imputing them would distort the target and introduce data leakage.

#### converting to numerical

```python
# convert numerical columns

cols_to_convert = ['temp','atemp','hum','windspeed','casual','registered','yr','mnth']
for col in cols_to_convert:
    df[col] = df[col].astype(float)
```

Converted numerical columns from object to float after cleaning invalid values.

```python
df.describe()
```

```python
df.shape
```

```python
df.isnull().sum()
```

#### Check Duplicates

```python
duplicated = df.duplicated().sum()
print(f"Duplicate rows found: {duplicated}")
```

#### filling missing values

```python
num_cols = ['temp','atemp','hum','windspeed']

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
cat_cols = ['season', 'yr', 'mnth', 'holiday', 'workingday', 'weathersit']

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

Used median for numerical features to handle skewness and mode for categorical features to preserve category distribution.

#### after filling missing values

```python
df.isnull().sum()
```

```python
# dropping instance column as it doesnt contribution to prediction variable

df.drop(['instant'], axis= 1, inplace=True)
```

```python
# converting to datetime
df['dteday'] = pd.to_datetime(df['dteday'], format='%d-%m-%Y')
```

Converted date column to datetime format to extract useful time-based features.

```python
df['day'] = df['dteday'].dt.day
df['month'] = df['dteday'].dt.month
df['year'] = df['dteday'].dt.year
df['day_of_week'] = df['dteday'].dt.dayofweek
```

Extracted meaningful components like day, month, and weekday to capture temporal patterns.

# Drop Unnecessary Columns

```python
df.drop(['dteday'], axis=1, inplace=True)
```

```python
df.drop(['month'], axis=1, inplace=True)
```

```python
df.drop(['yr'],axis=1, inplace =True)
```

```python
# Clean season
df['season'] = df['season'].astype(str).str.strip().str.title()
df['season'] = df['season'].replace({'Springer': 'Spring'})
```

```python
# dropping leakage collumns
df.drop(['casual', 'registered'], axis=1, inplace=True)
```

Dropped irrelevant columns like instant (index-like) and original date column after feature extraction.

# Feature Engineering

```python
# handling Time features
peak_list = []

for hour in df['hr']:
    if (hour >= 7 and hour <= 9) or (hour >= 17 and hour <= 19):
        peak_list.append(1)
    else:
        peak_list.append(0)

df['is_peak_hour'] = peak_list
```

Created peak hour feature to simplify time-based patterns and help model focus on high-demand periods.

```python
# creating weekend features

weekend_list = []

for day in df['weekday']:
    if day == 0 or day == 6:
        weekend_list.append(1)
    else:
        weekend_list.append(0)

df['is_weekend'] = weekend_list
```

Captures difference between weekday and weekend usage patterns.

```python
# bad weather feature

bad_weather_list = []

for w in df['weathersit']:
    if w == 3 or w == 4:
        bad_weather_list.append(1)
    else:
        bad_weather_list.append(0)

df['bad_weather'] = bad_weather_list
```

Simplified weather impact into good vs bad categories for easier model learning.
