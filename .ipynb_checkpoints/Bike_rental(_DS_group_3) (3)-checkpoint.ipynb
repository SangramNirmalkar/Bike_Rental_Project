#  Bike Sharing Rental — DS Group 3
> To predict hourly bike rental demand based on weather, season, and time factors. This enables bike-sharing companies to optimize bike availability, reduce waiting time, and improve customer satisfaction.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 100)
sns.set_theme(style="whitegrid") # Plot style
```

**1 - LOAD DATASET**

```python
df = pd.read_csv(r"C:\Users\admin\Downloads/Dataset.csv")
print(f"\nShape: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
df.head()
```

```python
df.tail()
```

Dataset has 17,379 hourly records from 2011-2012.
Several columns (temp, hum, yr, etc.) are stored as strings due to mixed '?' values - will be fixed in the cleaning step.

**2 - REPLACING INVALID '?' WITH NaN**

```python
for col in df.columns:
    mask = df[col].astype(str).str.strip() == '?'
    df.loc[mask, col] = np.nan

print(df.isnull().sum()[df.isnull().sum() > 0].to_string())
```

```python
df.head(10)
```

48 cells contained '?' used as a placeholder for unknown values.
Affected columns: season, yr, mnth, holiday, workingday,
weathersit, temp, atemp, hum, windspeed, casual, registered.
All replaced with NaN for proper handling.

**3 - DATA TYPE CONVERSION**

```python
df['dteday'] = pd.to_datetime(df['dteday'], dayfirst=True, errors='coerce')

num_cols = ['yr', 'mnth', 'temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered']
for c in num_cols:
    df[c] = pd.to_numeric(df[c], errors='coerce')

print(df.dtypes.to_string())
```

Date column parsed to datetime. Numeric columns (temp, hum,
windspeed, etc.) converted from string to float using pd.to_numeric().
Rows with non-convertible values are now NaN, handled in next step.

**4 - CHECK DUPLICATES**

```python
duplicated = df.duplicated().sum()
print(f"Duplicate rows found: {duplicated}")
```

No duplicate rows exist. Each row represents a unique hour
of a specific date, so zero duplicates is expected.


**5 - MISSING VALUE TREATMENT**

```python
print(f'Total number of missing values : {df.isnull().sum().sum()}')
df.isnull().sum()
```

```python
# Numeric - median imputation
for c in ['temp', 'atemp', 'hum', 'windspeed', 'yr', 'mnth', 'casual', 'registered']:
    df[c].fillna(df[c].median(), inplace=True)

# Categorical - mode imputation
for c in ['season', 'holiday', 'workingday', 'weathersit']:
    df[c].fillna(df[c].mode()[0], inplace=True)

# Fix integer types
df['yr']         = df['yr'].round().astype('Int64')
df['mnth']       = df['mnth'].round().astype('Int64')
df['casual']     = df['casual'].round().astype('Int64')
df['registered'] = df['registered'].round().astype('Int64')

print(f"Missing values after imputation: {df.isnull().sum().sum()}")
```

Numeric columns imputed with median (robust to outliers).
Categorical columns imputed with mode (most frequent category)      Dataset is now fully clean with 0 missing values.

**6 - DESCRIPTIVE STATISTICS**

```python
print(df[['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']].describe().round(3).to_string())
```

```python
print(f"\nCasual users     : {df['casual'].sum()/df['cnt'].sum()*100:.1f}% of total rentals")
print(f"Registered users : {df['registered'].sum()/df['cnt'].sum()*100:.1f}% of total rentals")
```

'cnt' (target) ranges 1-977 with mean ~189 and high std ~181,
indicating demand is highly variable across conditions.      Registered users dominate at 81% of all rentals.

**LABEL MAPPING (for readable plots)**

```python
month_map   = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',
               7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
weekday_map = {0:'Sun',1:'Mon',2:'Tue',3:'Wed',4:'Thu',5:'Fri',6:'Sat'}
season_order = ['Spring','Summer','Fall','Winter']

df['month_label']   = df['mnth'].map(month_map)
df['weekday_label'] = df['weekday'].map(weekday_map)
df['year_label']    = df['yr'].astype(float).round().astype('Int64').map({2011:'2011',2012:'2012'})
df['season_label']  = (df['season'].astype(str).str.strip().str.title()
                        .replace({'Springer':'Spring'}))
```

**7 - DATA VISUALIZATION**

Plot 1 - Distribution of Key Numeric Features

```python
feat_pairs = [
    ('cnt',        'Total Rentals (cnt)'),
    ('casual',     'Casual Users'),
    ('registered', 'Registered Users'),
    ('temp',       'Temperature (norm.)'),
    ('hum',        'Humidity (norm.)'),
    ('windspeed',  'Wind Speed (norm.)'),
]
colors = sns.color_palette("muted", len(feat_pairs))

f, a = plt.subplots(2, 3, figsize=(15, 8))
f.suptitle("Plot 1 - Distribution of Key Numeric Features", fontsize=13, fontweight='bold')

for ax, (col, title), color in zip(a.flat, feat_pairs, colors):
    ax.hist(df[col].dropna(), bins=40, color=color, edgecolor='white', linewidth=0.4)
    ax.set_title(title, fontsize=9)
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

plt.tight_layout()
```


* cnt and user counts are right-skewed — most hours have low rentals with occasional high-demand spikes.
* Temperature is fairly uniform across the year.
* Windspeed is heavily right-skewed; most hours have very low wind.






Plot 2 - Outlier Detection via Box Plots

```python
f, a = plt.subplots(1, 4, figsize=(16, 5))
f.suptitle("Plot 2 — Outlier Detection via Box Plots", fontsize=13, fontweight='bold')

for ax, col in zip(a, ['cnt', 'casual', 'registered', 'windspeed']):
    sns.boxplot(y=df[col], ax=ax, color='steelblue', width=0.4,
                flierprops=dict(marker='o', markersize=3, alpha=0.5))
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    n_out = ((df[col] < Q1-1.5*(Q3-Q1)) | (df[col] > Q3+1.5*(Q3-Q1))).sum()
    ax.set_title(col)
    ax.set_xlabel(f"Outliers: {n_out}", fontsize=8, color='red')

plt.tight_layout()
```

* High-end outliers exist in cnt, casual, and registered — these represent genuine peak-demand hours (events, good weather) not data errors.
* Outliers are retained as they carry important signal for the model.

Plot 3 - Correlation Heatmap

```python
num_df = df[['hr','mnth','weekday','temp','atemp','hum',
             'windspeed','casual','registered','cnt']].copy()
corr = num_df.corr()

f, a = plt.subplots(figsize=(11, 8))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, ax=a, annot_kws={'size': 9}, vmin=-1, vmax=1)
ax.set_title("Plot 3 - Correlation Matrix", fontsize=13, fontweight='bold')
plt.tight_layout()
```

* temperature has the strongest positive correlation with cnt (+0.40).
* Humidity is negatively correlated (−0.32) — wetter days = fewer riders.
* temp and atemp are nearly identical (r=0.99) we can drop one to avoid multicollinearity.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.impute import SimpleImputer

# --- 1. DATA PREPROCESSING ---
# Dropping columns not needed for prediction (id, date, and components of 'cnt')
# We drop 'casual' and 'registered' because they sum up to 'cnt' (the target)
cols_to_drop = ['instant', 'dteday', 'casual', 'registered']
df_model = df.drop(columns=cols_to_drop)

# Convert categorical columns to numeric using One-Hot Encoding
# This handles 'season', 'weathersit', 'holiday', and 'workingday'
df_model = pd.get_dummies(df_model, drop_first=True)

# Define Features (X) and Target (y)
X = df_model.drop('cnt', axis=1)
y = df_model['cnt']

# Handle any remaining NaN values (from the '?' replacement) with the median
imputer = SimpleImputer(strategy='median')
X_imputed = imputer.fit_transform(X)
y_imputed = y.fillna(y.median())

# --- 2. TRAIN/TEST SPLIT ---
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_imputed, test_size=0.2, random_state=42)

# --- 3. MODEL BUILDING (Random Forest) ---
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# --- 4. PREDICTION & EVALUATION ---
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("--- Model Performance ---")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-Squared (R2) Score: {r2:.4f}")


importances = pd.Series(model.feature_importances_, index=X.columns)
importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Features Driving Bike Demand")
plt.show()
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Define the grid of hyperparameters to search
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'bootstrap': [True, False]
}

# Use RandomizedSearchCV to find the best combination
rf_random = RandomizedSearchCV(estimator=model, param_distributions=param_grid, 
                               n_iter=10, cv=3, random_state=42, n_jobs=-1)
rf_random.fit(X_train, y_train)

print(f"Best Parameters: {rf_random.best_params_}")
best_model = rf_random.best_estimator_
```
